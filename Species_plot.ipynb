```python
from csv import reader
from datahandling import DataFile, alldatafiles, cuts, trim
from model_parameters import parameters
from curve_func_2 import model_curves
%pylab inline
#matplotlib.rc('font', family='serif')
#matplotlib.rc('font', serif='Cambria')
matplotlib.rc('font', size='15')
matplotlib.rc('lines', lw='2')
```

```python
with open('all_parameters_14.csv', 'rb') as csvfile:
    read = reader(csvfile)
    all_rows = []
    for row in read:
        all_rows.append(row)
```

```python
ini_val = []
data_set = 11
for n in all_rows[data_set][1:18]:
    ini_val.append(float(n))
```

```python
p = parameters(ini_val)
```

```python
files = alldatafiles()
time_data, temp_data, torque_data = DataFile(files[data_set - 1]).simple_data()

c = cuts(torque_data)
time_data = trim(time_data, c)
temp_data = trim(temp_data, c)
torque_data = trim(torque_data, c)
```

```python
curves = model_curves(p, time_data)
HCl, LDH, poly_act, radical, prim_stab, deg_poly, x_link, T, Tm, mu, torque = curves
```

```python
#figure(figsize=(6,5))
#plot(time_data, poly_act, '-.', label='poly_act')
#plot(time_data, radical, label='Radical')#, linewidth='3')#, alpha=0.7)
plot(time_data, prim_stab, label='Primary stabiliser', alpha=0.7, color='red')#, color='black')#, linewidth='3')#, alpha=0.7)
plot(time_data, LDH, label='LDH', alpha=0.7, color='blue')#, linewidth='3')#, alpha=0.7)
plot(time_data, HCl, label='HCl', alpha=0.7, color='green')#, linewidth='3')#, alpha=0.7)
plot(time_data, deg_poly, label='Degraded polymer', alpha=0.7, color='black')#, linewidth='3')#, alpha=0.7)
plot(time_data, x_link, label='Cross-link', alpha=0.7, color='magenta')#, linewidth='3')#, alpha=0.7)
xlabel('Time (min)')
ylabel('Concentration (phr)')
#legend(loc='upper left')
ylim([0, 1.4])
xlim([0, 15])
savefig('poster_species.pdf')
```
